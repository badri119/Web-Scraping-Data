```python
import tabula
file1 = ("/content/Sales.pdf")
table = tabula.read_pdf(file1,pages=1)
table[0]
```

```python
table2 = tabula.read_pdf(file1,pages=2)
table2[0]
```

```python
table3 = tabula.read_pdf(file1,pages=3)
table3[0]
```

```python
table4 = tabula.read_pdf(file1,pages=4)
table4[0]
```

```python
import pandas as pd


# Convert any non-DataFrame objects into DataFrame objects
table[0] = pd.DataFrame(table[0])
table2[0] = pd.DataFrame(table2[0])
table3[0] = pd.DataFrame(table3[0])
table4[0] = pd.DataFrame(table4[0])

# Concatenate all tables into one DataFrame
combined_table = pd.concat([table[0], table2[0], table3[0], table4[0]])

# Reset index to ensure continuous index values
combined_table.reset_index(drop=True, inplace=True)

# Display the combined table
print(combined_table)
```

```python
# Define the file path for the CSV file
csv_file_path = "/content/extracted_data.csv"

# Convert the DataFrame to a CSV file
combined_table.to_csv(csv_file_path, index=False)

print("Combined table has been successfully saved to CSV:", csv_file_path)
```
